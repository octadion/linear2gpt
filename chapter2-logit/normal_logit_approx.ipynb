```python
from scipy.stats import norm, logistic
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams.update({'font.size':12})
fig = plt.figure(figsize=(9, 6), dpi=100)
ax = fig.add_subplot(111)
x = np.linspace(-8, 4, 100)
alpha = 1.702
normal = norm.cdf(x)
logit = logistic.cdf(alpha * x)
ax.plot(x, normal,
        label='Standard Normal Distribution：'
            + r'$F(x) = \int_{-\infty}^{x}\frac{1}{\sqrt{2\pi}}{\rm e}^{-\frac{t^{2}}{2}}$'
            + r'${\rm d}t$')
ax.plot(x, logit, 'k-.',
        label=r'Best approximation to the logistic distribution：$F(x) = \frac{1}{1 + {\rm e}^{-1.702x}}$')
ax.set_xlabel('$x$')
ax.set_ylabel('Cumulative distribution function：$F(x)$')
legend = plt.legend(shadow=True, fontsize=16)
plt.savefig('normal_logit_approx.png', dpi=200)
plt.show()
```
